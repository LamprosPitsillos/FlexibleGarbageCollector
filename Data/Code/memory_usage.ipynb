```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
logs = ['counting_instant.csv', 'counting_late.csv']
labels = {
    'counting_instant.csv': 'Instantly destruct',
    'counting_late.csv': 'Lately destruct'
}
ax = None
for log_file in logs:
    test_df = pd.read_csv(f'../Logs/{log_file}')
    test_df['Timestamp'] = test_df['Dt (Î¼s)'].cumsum()
    ax = test_df.plot(
        x='Timestamp', y='Memory Used (bytes)',
        fontsize=16, figsize=(10, 4), title='Test', ax=ax,
        label=labels[log_file]
    )
ax.get_figure().savefig('../Plots/counting_memory_usage.png')
```
